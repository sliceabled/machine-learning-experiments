```python
import tensorflow as tf
mnist = tf.keras.datasets.mnist # 28x28 images of hand-written digits 0-9

(x_train, y_train),  (x_test, y_test) = mnist.load_data()

x_train = tf.keras.utils.normalize(x_train, axis=1)  # Acts as a sigmoid function - easy for learning
x_test = tf.keras.utils.normalize(x_train, axis=1)

model = tf.keras.models.Sequential() #sequential architecture for deep learning module 
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(128,  activation=tf.nn.relu)) 
model.add(tf.keras.layers.Dense(128,  activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(10,  activation=tf.nn.softmax)) 

model.compile(loss='sparse_categorical_crossentropy', 
              optimizer='adam', 
              metrics=['accuracy'])

model.fit(x_train,  y_train,  epochs=3)  # accuracy refers to amount of times number from 1-10 was guessed correctly.
```
